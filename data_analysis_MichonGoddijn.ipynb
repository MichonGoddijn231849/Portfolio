```python
from openpyxl import load_workbook


file_path = 'C:\\Users\\Administrator\\Downloads\\data_countries.xlsx' 
workbook = load_workbook(filename=file_path)
sheet = workbook.active

# Extracting data from the "E-waste_recycled_per_capita_Kg" column
e_waste_per_capita_data = []
for row in sheet.iter_rows(min_row=2, max_row=sheet.max_row, values_only=True):
    if row[3] is not None:
        e_waste_per_capita_data.append(row[3])

# Calculating and printing statistics
if e_waste_per_capita_data:
    # Mean
    mean_value = sum(e_waste_per_capita_data) / len(e_waste_per_capita_data)

    # Standard Deviation
    variance = sum((x - mean_value) ** 2 for x in e_waste_per_capita_data) / len(e_waste_per_capita_data)
    std_deviation = variance ** 0.5

    # Maximum and Minimum
    max_value = max(e_waste_per_capita_data)
    min_value = min(e_waste_per_capita_data)

    formatted_results_str = (f"Mean (kg): {mean_value:.2f}\n"
                             f"Standard Deviation (kg): {std_deviation:.2f}\n"
                             f"Maximum Value (kg): {max_value}\n"
                             f"Minimum Value (kg): {min_value}")

    print(formatted_results_str)
else:
    print("No valid data available in the selected column.")
```
