```python
from azure.ai.ml import MLClient, command, Input
from azure.identity import DefaultAzureCredential
from azure.ai.ml.constants import AssetTypes, InputOutputModes

# choose Fix A or Fix B:
ml_client = MLClient(
    credential          = DefaultAzureCredential(),
    subscription_id     = "0a94de80-6d3b-49f2-b3e9-ec5818862801",
    resource_group_name = "buas-y2",
    workspace_name      = "NLP1-2025",
)

data_asset = ml_client.data.get("data_pipeline", version="2")

test_job = command(
    display_name="test_download_data_pipeline_v2",
    command='bash -c "head -n 5 ${{inputs.data}}"',   # print first 5 lines
    inputs={
        "data": Input(
            path=data_asset.id,
            type=AssetTypes.URI_FILE,
            mode=InputOutputModes.DOWNLOAD
        )
    },
    environment="azureml:AzureML-sklearn-1.0-ubuntu20.04-py38-cpu@latest",
    compute="adsai-lambda-0"                 # adjust
)

run = ml_client.jobs.create_or_update(test_job)
ml_client.jobs.stream(run.name)
```
