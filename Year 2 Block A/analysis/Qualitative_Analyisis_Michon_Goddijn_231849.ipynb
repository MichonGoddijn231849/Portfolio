### **Qualitative Analysis Report: The Impact of Chatbot Accuracy on Customer Trust in Comparison to Traditional Human Assistance**

#### **1. Introduction**
This paper addresses the fact that since chatbot can be used by small and medium sized enterprises (SMEs) to replace the traditional customer service, this report aims to explore the percentage of chatbot accuracy, when used by SMEs, effects related to customer trust, compared to other forms of customer service, such as by humans. With increased automation in customer service, businesses have turned to chatbots more than ever and the role of accuracy in building or breaking trust is crucial for the success of high customer satisfaction.

The research question guiding this analysis is: *How does the accuracy of chatbots in SMEs compare to traditional human assistance in terms of customer trust?* Five participants who have experience talking to individuals (human agents) and chatbots in service settings were interviewed using five semi-structured interviews. The data was analysed using thematic analysis to uncover key themes which provide a rich understanding of the customer experience with chatbot technology.

#### **2. Methodology**
A **thematic analysis** approach was undertaken using the Braun & Clarke (2006) framework. The interview data were well analysed using this approach to identify the recurring themes. Thematic analysis is a flexible method that allows for in depth exploration of participants’ experiences and perceptions, making it an ideal method for qualitative research on customer trust.

The analysis proceeded through several stages:
1. **Familiarisation with the data**: The interview transcripts were reviewed in detail for pattern and important points of interest.

2. **Initial coding**: Accurary, trust, speed, and human intervention were some of the elements on which codes were generated from key phrases and recurring ideas.

3. **Searching for themes**: The codes were then combined into larger themes that encapsulate the main features of customer trust in regards to chatbot accuracy.

4. **Reviewing themes**: To make sure the themes really reflected the participants experience, the themes therefore were refined and reorganised.

5. **Defining themes**: The themes were clearly defined and the supporting quotes from the interviews were identified to substantiate the analysis.

#### **3. Results**

##### **Theme 1: Accuracy as the Core of Trust**
The interviews revealed one of the most critical factors, namely, **accuracy** of the chatbot responses. In each of the five interviews, participants emphasized that the accuracy of chatbot responses was the key element of trust in a chatbot, and in the company that it represented.

For instance, when a chatbot answered right and relevant to their problem, they noted that not only did they resolve their problems but also they developed trust in the business. Most interviewees held the expectation that chatbots would behave as reliably as human agents. Besides, the reality often missed that mark with chatbots unable to meet complex queries.
- **Interview 1**: "If they don’t give a good answer, then I don’t have trust in the service."
- **Interview 2**: "Accuracy is very important... I feel truly heard with the question I have."
- **Interview 4**: "It’s very important that those chatbots are good and that accurate answers are given to your questions."
- **Interview 5**: "The accuracy should be about 100% because otherwise, I don’t understand why you would use the chatbot."

Participants were forgiving of minor errors, but resented when chatbots consistently failed to return accurate or contextually relevant responses. Specifically, this erosion of trust was especially apparent when chatbots failed to understand the user’s query or responded in a generic fashion.

##### **Theme 2: Human Assistance as a Backup**
The second key theme was the need for **human fallback** when chatbots ran out of juice. Often, participants said chatbots worked well for simple questions, but struggled with more complex or nuanced questions. In instances of this sort, the ability to easily switch to a human agent was essential to retaining customer trust.

Human agents were found to be perceived as more reliable especially when dealing with complicated or sensitive issues, participants said. Experiences where chatbots failed to solve a problem and the user had to retype the same information over and over again contributed to this perception, and was often reinforced. We expected that, if the chatbot couldn’t answer a query, it would be able to sense that limitation and quickly transfer the customer to a human.
- **Interview 1**: "If the chatbot notices, 'Well, we're not getting any further,' and then immediately connects you to a human."
- **Interview 3**: "So the combination is nice, that you can first go through everything that can be handled by a chatbot. And if that doesn’t work, you go to a human."
- **Interview 4**: "I find it very nice if a chatbot can quickly connect me to a human agent from the company."
- **Interview 5**: "If you get a bot that doesn’t understand you for like six times in a row... it’s really important that it can transfer to a human quickly."

Alongside being simply a technical necessity, it was also indicative of the company’s desire to provide customers with **seamless transition** from chatbot to human agent. If a chatbot failed to pick up the conversation right away, this was perceived as poor service, subsequent trust with the company diminished.

##### **Theme 3: Frustration with Inaccurate or Generic Responses**
A second common theme was that participants had **frustration** in their chatbots giving generic or inaccurate responses. However, many interviewees pointed out that chatbots tended to misinterpret more complicated questions or offer canned, irrelevant answers. Usually, this experience was described as frustrating, especially when the customer is already dealing with an issue and wants to be resolved quickly.

When chatbots repeatedly gave incorrect or irrelevant answers, participants were particularly frustrated, and felt as if their time was being wasted.
- **Interview 2**: "Chatbots are a sort of automatically generated answers to standard questions... with humans, I feel more heard."
- **Interview 4**: "The chatbot just gave a completely different answer. That gave me a negative image of the company."
- **Interview 5**: "You just keep speaking to a bot that doesn’t understand you... and that could result in someone not buying products there again."

Not only was this frustration affecting the participants’ perception of the chatbot, but it also had a knock on effect on how they perceived the company. If the chatbot didn’t do a good job, it was bad for business (especially without an easy way out to human support).

##### **Theme 4: Expectations vs. Reality**
Participants held mixed expectations for chatbots and their met expectations had a direct impact on trust in the company. For many interviewees, they didn’t expect chatbots to be able to answer complex questions as well as human agents. But trust improved when chatbots did better than expected.

To the contrary, however, when participants predicted that their issue would be resolved by a chatbot, and the chatbot didn't, the response was disappointment and frustration. Several interviewees said that their faith in the chatbot was first shaken by its failure to live up to their expectations, especially on tasks they believed should be easy.
- **Interview 1**: "I had no expectation that they would do that, so yes, positively surprised."
- **Interview 3**: "If a chatbot can’t help you, then you don’t get any further, and that’s frustrating."
- **Interview 4**: "I think it’s better to have a human agent who responds to my difficult question than a chatbot."

This theme presents the point that one should not overestimate the potential performance of a chatbot. If the expectations created by sharing a real person would lead the customer to be surprised, chatbots brought customer trust while when the expectations are not met, the customer would be disappointed and trust would fall.

##### **Theme 5: Speed vs. Personalisation**
The final theme was on the **trade off between speed and personalisation**. While chatbots certainly sped things up and reduced the average handling time, participants always preferred human agents for more complex, or personalised questions. Participants believed chatbots would be useful for quickly answering basic questions, but for more in depth, context sensitive issues, they preferred talking to a human.

But several interviewees said that chatbots were quicker at processing routine tasks while human agents were better at delivering personalised solutions tailored to the customer’s needs. Stronger still was their preference for personalisation when participants were handling emotionally or sensitive issues.
- **Interview 2**: "With a live chat, I can go more specifically into the question, and I feel more heard."
- **Interview 3**: "The chatbot does it faster... but the human can handle more complex things."
- **Interview 5**: "You just want to speak to a human because that one can actually communicate like a normal person."

A **speed vs. depth** trade off proved to be an important determinant in whether participants preferred chatbots versus human agents. Participants still appreciated the personalised, context aware responses provided by human agents, even when they were praised for their efficiency.

Clearly, these factors are important for how customers judge chatbot performance, and so SMEs should focus on these aspects to build more trust and satisfaction. #### **4. Conclusion**
Analysis shows that customer trust in SMEs is dependent on the accuracy of chatbot. When chatbots get it right, and help customers to feel understood and cared for, then they help to shape positive customer experiences and build trust. Yet, any incorrect or generic answer will soon make the chatbot and company lose trust. Especially when the chatbot isn’t equipped to get complex queries, or when the client demands greater personalisation, human fallback is crucial.

While chatbots are known to be quite fast and efficient, they are often considered as 'thin' and not personal enough, and that's exactly what a lot of the buyers are looking for in intricate interactions. Consequently, SMEs should concentrate on increasing the accuracy of chatbot, managing customer’s expectations, and passing the baton to human support when necessary. This balanced approach will ensure that customer trust is maintained and the overall satisfaction with the service will be improved.
